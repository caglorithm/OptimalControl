```python
#!/usr/bin/env python

import sys
import os
absFilePath = os.path.abspath('attractor_switching.py')
fileDir = os.path.dirname(absFilePath)
parentDir = os.path.dirname(fileDir)

# change working directory to the project file of codes, use parentDir else
sys.path.append(parentDir+'/code')


import functions_noisy as functions
import other_functions
from FRoptimize_noisy import FR_algorithm
import scipy.sparse as sp
import numpy as np
import random
import matplotlib
import pickle
import matplotlib.pyplot as plt
import random
import scipy as sc
import scipy.signal as ss
from pylab import figure, cm
from scipy import signal


#################################################################################################################################
#FUNCTIONS #####################################################################################################################
#################################################################################################################################

def file_name(parentDir,I_p,I_e,I_s,T,eta,run,bifurcation):
    folder=parentDir+'/results/synchronization_'+bifurcation
        
    return folder+'/aperiodic_syncI_p'+str(round(I_p,10))+'I_e'+str(round(I_e,10))+'I_s'+str(round(I_s,10))+'T'+str(int(T))+'noisevar'+str(round(eta,10))+'run'+str(run)+'.0_'


def readfile(filename):
    with open(filename, 'rb') as pickle_file:
        content = pickle.load(pickle_file)
    warnflag=content['warnflag']
    x0=content['optimal_control']
    ic=content['ic']

    mu=content['external_current']
    sigma=content['coupling_strength']
    return x0,ic,mu,sigma,warnflag
        
    
#################################################################################################################################
#PARAMETERS #####################################################################################################################
#################################################################################################################################



############## COUPLING #########################################

A = np.load(parentDir+"/data/SC_m12av_voxNorm_th20.npy")
N = A.shape[0]
degree=np.sum(A,axis=0)
d=2


##################### FHN NODES #################################
'''
du/dt = -alpha u^3 + beta u^2 - gamma u - w + mu
dw/dt = 1/tau (u + delta  - epsilon w)
'''
alpha=3.
beta=4.
gamma=3./2
tau=20.
delta=0
epsilon=0.5

############# FUNCTIONAL ###################
I_p= 0.1
I_e= 1.0
I_s_ar= np.arange(0.0,100.0,10.0)
small=1e-5
limin= 10.0#float(sys.argv[4])#*a_steps
T= 500
T_plot=T+150
dt=0.1
sqrt_dt=dt**(1/2)
end_T=int(T/dt)
end_T_plot=end_T+1500
eta_ar=np.array([0.0,0.012,0.024])
bifurcation_ar=np.array(['low','high'])
runs=5
runs_ar=np.arange(runs)
runs_noise=100
runs_noise_ar=np.arange(runs_noise)
seed=50
```

```python
#initialize
cross_correlation=np.zeros((bifurcation_ar.shape[0],eta_ar.shape[0],I_s_ar.shape[0],runs,runs_noise))
uncontrolled_cross_correlation=np.zeros((bifurcation_ar.shape[0],eta_ar.shape[0]))

#for additional figs
energy_total=np.zeros((bifurcation_ar.shape[0],eta_ar.shape[0],I_s_ar.shape[0],runs))
controlled=np.zeros((bifurcation_ar.shape[0],eta_ar.shape[0],I_s_ar.shape[0],runs))

for e,eta in enumerate(eta_ar):
    ####INITIALIZE NOISE####
    noise =  (eta/sqrt_dt)*np.random.normal(size=(runs_noise,end_T,N))

    for b,bifurcation in enumerate(bifurcation_ar):
        for i,I_s in enumerate(I_s_ar):
            for run in runs_ar:
                filename=file_name(parentDir,I_p,I_e,I_s,T,eta,run,bifurcation)

                x0,ic,mu,sigma,warnflag=readfile(filename)
                print(eta,I_s,warnflag)

                
                #### for additional figs
                energy_total[b,e,i,run]=np.sum(x0**2)*dt

                controlsum= np.sum(x0**2, axis=0)**(1.0 / 2.0)           
                for j in np.arange(0,N):
                    if controlsum[j]>1e-5:
                        controlled[b,e,i,run]+=1
                ####
                
                
                r_noise=0
                while r_noise<runs_noise:
                    state1=other_functions.runge_kutta_noisy(other_functions.FHN_noisy,ic, x0, end_T , d , dt , N , alpha , beta, gamma, delta, epsilon, tau, mu ,sigma, A,noise[r_noise])
                    cross_correlation[b,e,i,run,r_noise]=np.mean(other_functions.cross_correlation_matrix(state1[:,0,:] , end_T , N))
                    if I_s==0.0:
                        state0=other_functions.runge_kutta_noisy(other_functions.FHN_noisy,ic, np.zeros((end_T,N)), end_T , d , dt , N , alpha , beta, gamma, delta, epsilon, tau, mu ,sigma, A,noise[r_noise])
                        uncontrolled_cross_correlation[b,e]+=np.mean(other_functions.cross_correlation_matrix(state0[:,0,:] , end_T , N))/(runs)
                    if eta==0.0:
                        r_noise=runs_noise
                        cross_correlation[b,e,i,run,:]=cross_correlation[b,e,i,run,0]*np.ones(runs_noise)
                    else:
                        r_noise+=1

```

```python
uncontrolled_cross_correlation[0,0]=uncontrolled_cross_correlation[0,0]/(runs)
uncontrolled_cross_correlation[1,0]=uncontrolled_cross_correlation[1,0]/(runs)
lw=2
fs=30

x=I_s_ar

cross_correlation = np.resize(cross_correlation, (bifurcation_ar.shape[0],eta_ar.shape[0],I_s_ar.shape[0],runs*runs_noise))
y=np.mean(cross_correlation, axis=3)
yerror=np.std(cross_correlation, axis=3)

cols=['black','green','red']
labels=['$\eta=0.0$','$\eta=0.012$','$\eta=0.024$']

for b,bifurcation in enumerate(bifurcation_ar):
    fig, (ax) = plt.subplots(nrows=1, sharex=True, figsize=(8,5))
    kwargs = dict(ecolor='k', color='k', capsize=2,
                  elinewidth=1.1, linewidth=0.6, ms=11)
    for e,eta in enumerate(eta_ar):
        ax.errorbar(x,y[b,e] ,yerr=yerror[b,e], fmt='o', mfc=cols[e],label=labels[e], **kwargs)
        ax.plot(np.arange(0,95),np.ones((95))*uncontrolled_cross_correlation[b,e],color=cols[e])

    ax.set_xlabel('sparsity $I_s$',size=fs)
    ax.set_ylabel('cross-correlation',size=fs)
    #ax.set_title('mean and std of control energy over 10 runs',size=fs, pad=20)
    ax.tick_params(labelsize=fs)
    ax.grid(True)
    ax.set_ylim(0,np.max(y)+np.max(yerror))
    ax.legend(loc='lower right', frameon=True,fontsize=fs)
    ax.legend(loc='upper right', ncol=1, frameon=True,fontsize=fs-9)


    plt.savefig(parentDir+'/figures/fig_paper/synchronize/notebook_3_sparse/sparse_'+bifurcation+'.png', bbox_inches='tight')   

    plt.show()
```
