```python
#!/usr/bin/env python

import sys
import os
absFilePath = os.path.abspath('attractor_switching.py')
fileDir = os.path.dirname(absFilePath)
parentDir = os.path.dirname(fileDir)

# change working directory to the project file of codes, use parentDir else
sys.path.append(parentDir+'/code')


import functions_noisy as functions
import other_functions
from FRoptimize_noisy import FR_algorithm
import scipy.sparse as sp
import numpy as np
import random
import matplotlib
import pickle
import matplotlib.pyplot as plt
import random
import scipy as sc
import scipy.signal as ss
from pylab import figure, cm
from matplotlib.colors import LogNorm
from matplotlib.lines import Line2D
from scipy import signal

import scipy as sc
import scipy.linalg as la
import scipy.stats as st

#################################################################################
#---FUNCTIONS FOR LINEAR CONTR: MEASURES (MODAL AND AVERAGE CONTROLLABILITY) ---#

def calc_mod_control(A):
    '''
    Calculate modal controllability for a given adjacency matrix
    '''
    evals, evecs = np.linalg.eig(A)
    modal_c = np.zeros(94)
    for i in range(A.shape[0]):
        for j in range(A.shape[0]):
            modal_c[i] += (1-evals[j]**2)*evecs[i,j]**2
    return modal_c


def pairmul(v):
    vv = np.zeros((N,N))
    for i in np.arange(N):
        for j in np.arange(N):
            vv[i,j]=v[i]*v[j]
    return vv

def calc_ave_control(A):
    '''
    calculate Gramian, average controllability, 
    based on https://complexsystemsupenn.com/s/controllability_code-smb8.zip
    '''
    N = A.shape[0]
    T,Z=la.schur(A)
    midMat= (Z**2).transpose()
    v=np.diag(T)
    vv = np.zeros((N,N))
    for i in np.arange(N):
        for j in np.arange(N):
            vv[i,j]=v[i]*v[j]
    P= np.repeat(np.diag(1-vv),N).reshape(N,N)
    average_c=np.sum(midMat/P , axis=0).transpose()
    return average_c


#########################################################################
###################---FUNCTIONS FOR READING FILES---#####################

def file_name(parentDir,bifurcation,switchfromto,I_p,I_e,I_s,T,phase):
    folder=parentDir+'/results/state_switching'
    return folder+'/trans'+bifurcation+switchfromto+'I_p'+str(round(I_p,10))+'I_e'+str(round(I_e,10))+'I_s'+str(round(I_s,10))+'T'+str(int(T))+'phase'+str(int(phase))+'_'
def readfile(filename):
    with open(filename, 'rb') as pickle_file:
        content = pickle.load(pickle_file)
    warnflag=content['warnflag']
    control=content['optimal_control']
    
    return control,warnflag
        
def find_inis(times_before,ic, d , dt , N , alpha , beta, gamma, delta, epsilon, tau, mu ,sigma, A):
    end_T_b1=20000
    end_T_b=2000
    stateb=other_functions.runge_kutta(other_functions.FHN,ic, np.zeros((end_T_b1,N)), end_T_b1 , d , dt , N , alpha , beta, gamma, delta, epsilon, tau, mu ,sigma, A)

    if stateb[0,0,22]<stateb[1,0,22]:
        for t in np.arange(end_T_b,end_T_b1):
            if stateb[t,0,22]<ic[0,22] and stateb[t+1,0,22]>ic[0,22] :
                ic1=stateb[t-times_before]
                break
    elif stateb[0,0,22]>=stateb[1,0,22]:
        for t in np.arange(end_T_b,end_T_b1):
            if stateb[t,0,22]>ic[0,22] and stateb[t+1,0,22]<ic[0,22] :
                ic1=stateb[t-times_before]
                break
    return ic1      



#################################################################################################################################
#--PARAMETERS-- #####################################################################################################################
#################################################################################################################################



############## COUPLING #########################################

A = np.load(parentDir+"/data/SC_m12av_voxNorm_th20.npy")
N = A.shape[0]
degree=np.sum(A,axis=0)
d=2

##################### PARAMETERS OF FHN NODES #################################
'''
du/dt = -alpha u^3 + beta u^2 - gamma u - w + mu
dw/dt = 1/tau (u + delta  - epsilon w)
'''
alpha=3.
beta=4.
gamma=3./2
tau=20.
delta=0
epsilon=0.5

####### PARAMETERS COST FUNCTIONAL AND TIMES ############################
#########################################################################
I_e= 1.0
I_s= 0.0
limin= 10.0
T= 400
dt=0.1
end_T=int(T/dt)
```

```python
##### LOAD DATA ####################################
####################################################

bifurcation_ar=np.array(['low', 'high'])
switchfromto_ar=np.array(['SB', 'BS'])

no_examples=4

energy=np.zeros((bifurcation_ar.shape[0],switchfromto_ar.shape[0],N))


for b,bifurcation in enumerate(bifurcation_ar):
    for s,switchfromto in enumerate(switchfromto_ar):
        
        if bifurcation=='high':
            I_p= 7e-5
            if switchfromto=='SB':
                phase=354
            elif switchfromto=='BS':
                phase=192
        elif bifurcation=='low':
            I_p= 0.0005
            phase=0

        filename=file_name(parentDir,bifurcation,switchfromto,I_p,I_e,I_s,T,phase)
        #print(filename)

        control,warnflag=readfile(filename)
        #print(bifurcation,switchfromto,warnflag)

        energy[b,s]=np.sum(control**2,axis=0)*dt
```

```python
#Linear control measures
#for stable linear dynamics 
normalize=2*np.max(np.linalg.eig(A)[0])
modal_c=calc_mod_control(A/normalize)
average_c=calc_ave_control(A/normalize)
linear_measures=np.array([average_c,modal_c])
linear_measures_ar=np.array(['average controllablity','modal controllability'])
```

```python

cols=['black','red']
markerSize=70
fs=30
plt.rc('font', size=fs)

minn=np.min(average_c)
maxx=np.max(average_c)
add=(maxx-minn)/25
min_ave=minn-add
max_ave=maxx+add
minn=np.min(modal_c)
maxx=np.max(modal_c)
add=(maxx-minn)/25
min_mod=minn-add
max_mod=maxx+add
mi=np.array([min_ave,min_mod])
ma=np.array([max_ave,max_mod])


for b,bifurcation in enumerate(bifurcation_ar):
    for m,measure in enumerate(linear_measures_ar):
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,6))
        ax2 = ax.twinx()
        s=0
        ax.scatter(linear_measures[m],energy[b,s], marker="o", color=cols[s],s=markerSize+25, edgecolor='black', linewidth='1')
        ax.set_ylabel('control energy \n low to high',size=fs,color=cols[0])
        ax.set_xlabel(str(linear_measures_ar[m]),size=fs)
 #       ax.set_title(bifurcation+' bifurcation',size=fs, pad=40)
        ax.tick_params(labelsize=fs)
        ax.set_ylim(0,np.max(energy[b,s])+(np.max(energy[b,s])-np.min(energy[b,0]))/25)
        ax.set_xlim(mi[m],ma[m])
        ax.tick_params(axis='y', labelcolor=cols[0],labelsize=fs)
        ax.ticklabel_format(style='sci',scilimits=(0,0),axis='y')
        ax.xaxis.major.formatter._useMathText = True

        s=1
        ax2.scatter(linear_measures[m],energy[b,s], marker="D", color=cols[s],s=markerSize, edgecolor='black', linewidth='1')
        ax2.set_ylabel('control energy \n high to low',size=fs,color=cols[1])
        ax2.set_ylim(0,np.max(energy[b,s])+(np.max(energy[b,s])-np.min(energy[b,0]))/25)
        ax2.set_xlim(mi[m],ma[m])
        ax2.tick_params(axis='y', labelcolor=cols[1],labelsize=fs)
        ax2.ticklabel_format(style='sci',scilimits=(0,0),axis='y')
        ax2.xaxis.major.formatter._useMathText = True
        ax.grid(True) 
        
        plt.savefig(parentDir+'/figures/fig_paper/linear_vs_nonlinear/notebook_1_switching/ln_switch_fig_'+str(m)+str(b)+'.png', bbox_inches='tight')   
        plt.show()
```
