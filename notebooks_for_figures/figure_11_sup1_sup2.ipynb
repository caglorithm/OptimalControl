```python
#!/usr/bin/env python

import sys
import os
absFilePath = os.path.abspath('attractor_switching.py')
fileDir = os.path.dirname(absFilePath)
parentDir = os.path.dirname(fileDir)

# change working directory to the project file of codes, use parentDir else
sys.path.append(parentDir+'/src')

import numpy as np
import matplotlib
import pickle
import matplotlib.pyplot as plt
from pylab import figure, cm

```

```python
filename=fileDir+'/data/data_figure_11'
with open(filename, 'rb') as pickle_file:
    data = pickle.load(pickle_file)
print(data.keys())
uncontrolled_cross_correlation=data['uncontrolled_cross_correlation']
mean_cross_correaltion=data['mean_cross_correaltion']
std_cross_correlation=data['std_cross_correlation']
values_I_s=data['values_I_s']
# for additional backup plots
mean_perc_of_contr_nodes=data['mean_perc_of_contr_nodes']/100 #all parameters same as above. mean percentage of controlled nodes
std_perc_of_contr_nodes=data['std_perc_of_contr_nodes']/100
mean_total_control_energy=data['mean_total_control_energy'] # all parameters same as above. mean control energy on all nodes
std_total_control_energy=data['std_total_control_energy']
figures=['11a','11b']
```

```python

eta_ar=np.array([0.0,0.012,0.024])
bifurcation_ar=np.array(['low','high'])

lw=2
fs=30

x=values_I_s
y=mean_cross_correaltion
yerror=std_cross_correlation

figures=(['11a','11b'])
cols=['black','green','red']
labels=['$\eta=0.0$','$\eta=0.012$','$\eta=0.024$']

for b,bifurcation in enumerate(bifurcation_ar):
    fig, (ax) = plt.subplots(nrows=1, sharex=True, figsize=(8,5))
    kwargs = dict(ecolor='k', color='k', capsize=2,
                  elinewidth=1.1, linewidth=0.6, ms=11)
    for e,eta in enumerate(eta_ar):
        ax.errorbar(x[:8],y[b,e,:8] ,yerr=yerror[b,e,:8], fmt='o', mfc=cols[e],label=labels[e], **kwargs)
        ax.plot(np.arange(0,75),np.ones((75))*uncontrolled_cross_correlation[b,e],color=cols[e])

    ax.set_xlabel('sparsity $I_s$',size=fs)
    ax.set_ylabel('cross-correlation',size=fs)
    #ax.set_title('mean and std of control energy over 10 runs',size=fs, pad=20)
    ax.tick_params(labelsize=fs)
    ax.grid(True)
    ax.set_ylim(0,np.max(y)+np.max(yerror))
    ax.legend(loc='lower right', frameon=True,fontsize=fs)
    ax.legend(loc='upper right', ncol=1, frameon=True,fontsize=fs-9)


#    plt.savefig(parentDir+'/figures/fig_paper/synchronize/notebook_3_sparse/sparse_'+bifurcation+'.png', bbox_inches='tight')   
    print('Figure',figures[b])
    plt.savefig(fileDir+'/figures/figure_'+figures[b]+'.png', bbox_inches='tight') 
    plt.show()
```
