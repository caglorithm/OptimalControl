```python
#!/usr/bin/env python

import sys
import os
absFilePath = os.path.abspath('attractor_switching.py')
fileDir = os.path.dirname(absFilePath)
parentDir = os.path.dirname(fileDir)

# change working directory to the project file of codes, use parentDir else
sys.path.append(parentDir+'/src')

import numpy as np
import matplotlib
import pickle
import matplotlib.pyplot as plt
from pylab import figure, cm
```

```python
filename=fileDir+'/data/data_figure_12'
with open(filename, 'rb') as pickle_file:
    data = pickle.load(pickle_file)
bifurcation_ar=data['bifurcation_ar']
energy=data['control_energy']
average_c=data['average_controllability']
modal_c=data['modal_controllability']
figures=['12a','12b','12c','12d']
```

```python

cols=['black','red']
markerSize=70
fs=30
plt.rc('font', size=fs)

minn=np.min(average_c)
maxx=np.max(average_c)
add=(maxx-minn)/25
min_ave=minn-add
max_ave=maxx+add


for b,bifurcation in enumerate(bifurcation_ar):
    print('Figure',figures[b])
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,6))
    ax2 = ax.twinx()
    s=0
    ax.scatter(average_c,energy[b,s], marker="o", color=cols[s],s=markerSize+25, edgecolor='black', linewidth='1')
    ax.set_ylabel('control energy \n low to high',size=fs,color=cols[0])
    ax.set_xlabel('average controllability',size=fs)
#       ax.set_title(bifurcation+' bifurcation',size=fs, pad=40)
    ax.tick_params(labelsize=fs)
    ax.set_ylim(0,np.max(energy[b,s])+(np.max(energy[b,s])-np.min(energy[b,0]))/25)
    ax2.set_xlim(min_ave,max_ave)
    ax.tick_params(axis='y', labelcolor=cols[0],labelsize=fs)
    ax.ticklabel_format(style='sci',scilimits=(0,0),axis='y')
    ax.xaxis.major.formatter._useMathText = True

    s=1
    ax2.scatter(average_c,energy[b,s], marker="D", color=cols[s],s=markerSize, edgecolor='black', linewidth='1')
    ax2.set_ylabel('control energy \n high to low',size=fs,color=cols[1])
    ax2.set_ylim(0,np.max(energy[b,s])+(np.max(energy[b,s])-np.min(energy[b,0]))/25)
    ax2.set_xlim(min_ave,max_ave)
    ax2.tick_params(axis='y', labelcolor=cols[1],labelsize=fs)
    ax2.ticklabel_format(style='sci',scilimits=(0,0),axis='y')
    ax2.xaxis.major.formatter._useMathText = True
    ax.grid(True) 

    plt.savefig(fileDir+'/figures/figure_'+figures[b]+'.png', bbox_inches='tight') 
    plt.show()
```
